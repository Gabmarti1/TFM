```python
# we import modules from different libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#we are splitting the data to train and test the model
#we select the size of each part, and we established random state and stratify options. Tghe first one allows us to compare the different tryings or models
#the partititon will be always the same. The second one allows us to maintain the same proportioin of labels in both parts
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y, test_size = 0.30, random_state=125,stratify = Y)
```

```python
# we import different standaritzation models
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler

sc = StandardScaler()
Rsc=RobustScaler()
MMc=MinMaxScaler()
#Xtrain = sc.fit_transform(Xtrain)
#Xtest = sc.transform(Xtest)
Xtrain = Rsc.fit_transform(Xtrain)
Xtest = Rsc.transform(Xtest)
#Xtrain = MMc.fit_transform(Xtrain)
#Xtest = MMc.transform(Xtest)
```

```python
#we established all the differents possible values of each parameter
param_grid={'C' : [10000],'max_iter':[100000,1000000]
}

#we established all the differents possible values of each parameter
# param_grid={'C' : [0.1,0.01,1,10,100,1000,10000,100000],'max_iter':[10,100,1000,10000,100000,1000000]
# }
```

```python
#we are adjusting the best parametrs for our model
clf = GridSearchCV(LogisticRegression(), param_grid,cv=10)
```

```python
#here we train the model
clf.fit(Xtrain,Ytrain)
```

```python
#Obtain in the variable model the best logistic regression classifier

model = clf.best_estimator_
print(model)
print(model.coef_)


```

```python
# we are looking for the best cv value for a given model
dicc = {}
for i in range(2,22):
  scores1 = cross_val_score(LogisticRegression(), Xtrain, Ytrain, cv=i, scoring='accuracy')
  dicc[i] = np.mean(scores1)
import operator
best_CV = max(dicc, key = dicc.get)
```
