```python
def get_dicc(file,excluidos,just_in_clinical,i):
  import re
  n = 1
  cont = 0
  lab_sexo={}
  for lines in open(file):
    if n == 1:
      n = 0
      continue
    cont += 1
    if cont >= 112:
      continue
    lines = lines.strip('\n').split(';')
    id = lines[0]
    ident = re.split('[A-Z]+',id)
    print(ident)
    label = lines[i]
    if ident[1] in excluidos or ident[1] in just_in_clinical:
      continue
    lab_sexo[ident[1]]= label
  return lab_sexo
```

```python
def column(file,dicc):
  new_column = []
  for line1 in open(file):
    line1 = line1.strip('\n').split(';')
    name = line1[0].split('.')[0]
    for entry in dicc.keys():
      if name == entry:
        new_column.append(dicc[entry])
  return new_column
```
